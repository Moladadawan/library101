```python
# Test for plotly data
import plotly.express as px
import pandas as pd
df = pd.read_csv("gpm.csv")
fig = px.violin(df, y="pop", x="continent", color="continent", box=True, points="all", hover_data=df.columns)
fig.show()
```

```python
import plotly.express as px
import pandas as pd
df = pd.read_csv("test.csv")
fig = px.violin(df, y="Value", x="Item", color="Area", box=True, points="all", hover_data=df.columns,title='Emission of Carbondioxide and Methane by South Asia')
fig.show()
```

```python
import plotly.express as px
import pandas as pd
df = pd.read_csv("test.csv")
fig = px.scatter(df, x="Area", y="Value", color="Item",title='Emission of Carbondioxide and Methane by South Asia')
fig.show()
```

```python
import plotly.express as px
import pandas as pd
df = pd.read_csv("test.csv")
fig = px.bar(df, x="Area", y="Value", color="Item", barmode="group",title='Emission of Carbondioxide and Methane by South Asia')
fig.show()
```

```python
import plotly.express as px
import pandas as pd
df = df = pd.read_csv("asiae.csv")
# df.loc[df['Value'] < 2.e6, 'Area'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='Value', names='Area', title='Emission of Carbondioxide and Methane by South Asia')
fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder().query("year == 2007")
fig = px.sunburst(df, path=['continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'], title = "Emission of Carbondioxide and Methane by World")
fig.show()
```
